```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
input = [line.split(' ') for line in open('composite_quasar_spectrum.txt')]
x = np.zeros(len(input))
v = np.zeros(len(input))
for j in xrange(len(input)):
    temp = input[j][0]
    x[j] = float(temp)
    temp2 = input[j][2]
    v[j] = float(temp2)

plt.plot(x,v)
```

```python
file1 = [line.split() for line in open('sdss_filters/u_SDSS.res')]
file2 = [line.split() for line in open('sdss_filters/g_SDSS.res')]
file3 = [line.split() for line in open('sdss_filters/r_SDSS.res')]
file4 = [line.split() for line in open('sdss_filters/i_SDSS.res')]
file5 = [line.split() for line in open('sdss_filters/z_SDSS.res')]
u, uwav, g, gwav, r, rwav, i, iwav, z, zwav = np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)), np.zeros(len(file1)) 
```

```python
for y in xrange(len(file1)):
    temp1 = file1[y][1]
    u[y] = float(temp1)
    wtemp1 = file1[y][0]
    uwav[y] = float(wtemp1)
    temp2 = file2[y][1]
    g[y] = float(temp2)
    wtemp2 = file2[y][0]
    gwav[y] = float(wtemp2)
    temp3 = file3[y][1]
    r[y] = float(temp3)
    wtemp3 = file3[y][0]
    rwav[y] = float(wtemp3)
    temp4 = file4[y][1]
    i[y] = float(temp4)
    wtemp4 = file4[y][0]
    iwav[y] = float(wtemp4)
    temp5 = file5[y][1]
    z[y] = float(temp5)
    wtemp5 = file5[y][0]
    zwav[y] = float(wtemp5)
```

```python
plt.hold(True)
plt.plot(uwav, u)
plt.plot(gwav, g)
plt.plot(rwav, r)
plt.plot(iwav, i)
plt.plot(zwav, z)
plt.plot(x,v/10)
```

```python
unew = np.convolve(u,v, mode='same')
gnew = np.convolve(g,v, mode='same')
rnew = np.convolve(r,v, mode='same')
inew = np.convolve(i,v, mode='same')
znew = np.convolve(z,v, mode='same')
plt.hold(True)
plt.plot(unew, c='blue')
plt.plot(gnew)
plt.plot(rnew, c='black')
plt.plot(inew)
plt.plot(znew,c='red')
plt.plot(v)
```
