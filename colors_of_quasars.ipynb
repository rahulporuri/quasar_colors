```python
data = pd.read_csv('quasar_colors_rahul.poruri.csv')

redshift = []
u, g, r, i, z = [], [], [], [], []
u_g, g_r, r_i, i_z = [], [], [], []
```

```python
redshift = data['z']
u, g, r, i, z = data['modelMag_u'], data['modelMag_g'], data['modelMag_r'], data['modelMag_i'], data['modelMag_z']
u_g, g_r, r_i, i_z = data['modelMag_u'] - data['modelMag_g'], data['modelMag_g'] - data['modelMag_r'], data['modelMag_r'] - data['modelMag_i'], data['modelMag_i'] - data['modelMag_z']
```

```python
plt.hist(redshift, bins=50)
```

```python
plt.ylim(-4,5)
plt.scatter(redshift, u_g, s=4)
plt.ylabel('u-g')
plt.xlabel('redshift')
```

```python
%matplotlib inline
```

```python
plt.hold(True)

plt.subplot(221)
plt.ylim(-4,5)
plt.scatter(redshift, u_g, s=4)
plt.ylabel('u-g')

plt.subplot(222)
plt.ylim(-4,5)
plt.scatter(redshift, g_r, s=4)
plt.ylabel('g-r')

plt.subplot(223)
plt.ylim(-4,5)
plt.scatter(redshift, r_i, s=4)
plt.ylabel('r-i')
plt.xlabel('redshift')

plt.subplot(224)
plt.ylim(-4,5)
plt.scatter(redshift, i_z, s=4)
plt.ylabel('i-z')
plt.xlabel('redshift')

plt.show()
```
