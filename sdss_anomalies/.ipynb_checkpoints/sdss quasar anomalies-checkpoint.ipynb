SQL Query used to retrieve data for quasars in the vertical features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
first_set = pd.read_csv('first_set.csv')
```

```python
first_set
```

```python
first_set_ra = np.zeros(len(first_set))
first_set_dec = np.zeros(len(first_set))
first_set_ra = first_set['ra']
first_set_dec = first_set['dec']
first_set_ra = (first_set_ra-180)*np.pi/180
first_set_dec = first_set_dec*np.pi/180
```

```python
plt.subplot(111,projection='aitoff')
plt.scatter(first_set_ra,first_set_dec)
plt.grid(True)
```

```python
first_set_u, first_set_g, first_set_r, first_set_i, first_set_z = np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set))
first_set_u_g, first_set_g_r, first_set_r_i, first_set_i_z = np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set)), np.zeros(len(first_set))
first_set_u, first_set_g, first_set_r, first_set_i, first_set_z = first_set['modelMag_u'], first_set['modelMag_g'], first_set['modelMag_r'], first_set['modelMag_i'], first_set['modelMag_z']
first_set_u_g, first_set_g_r, first_set_r_i, first_set_i_z = first_set_u - first_set_g, first_set_g - first_set_r, first_set_r - first_set_i, first_set_i - first_set_z
```

```python
redshift = np.zeros(len(first_set))
redshift = first_set['z']
```

```python
plt.subplot(221)
plt.scatter(redshift,first_set_u_g)
plt.ylabel('u-g color')
plt.subplot(222)
plt.scatter(redshift,first_set_g_r)
plt.ylabel('g-r color')
plt.subplot(223)
plt.scatter(redshift,first_set_r_i)
plt.ylabel('r-i color')
plt.xlabel('redshift')
plt.subplot(224)
plt.scatter(redshift,first_set_i_z)
plt.ylabel('i-z color')
plt.xlabel('redshift')
```

```python
data = pd.read_csv('../quasar_colors_rahul.poruri.csv')

redshift = np.zeros(len(data))
u, g, r, i, z = np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data))
u_g, g_r, r_i, i_z = np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data))

redshift = data['z']
u, g, r, i, z = data['modelMag_u'], data['modelMag_g'], data['modelMag_r'], data['modelMag_i'], data['modelMag_z']

for item in np.where(u==-9999.0):
    u[item] = np.mean(u)
    
for item in np.where(g==-9999.0):
    g[item] = np.mean(g)

for item in np.where(r==-9999.0):
    r[item] = np.mean(r)

for item in np.where(i==-9999.0):
    i[item] = np.mean(i)
    
for item in np.where(z==-9999.0):
    z[item] = np.mean(z)
    
u_g, g_r, r_i, i_z = data['modelMag_u'] - data['modelMag_g'], data['modelMag_g'] - data['modelMag_r'], data['modelMag_r'] - data['modelMag_i'], data['modelMag_i'] - data['modelMag_z']
```

```python
plt.subplot(221)
plt.hold(True)
plt.scatter(u_g, g_r, s=2)
plt.scatter(first_set_u_g, first_set_g_r, s=10, c='r')
plt.ylabel('u-g')

plt.subplot(222)
plt.hold(True)
plt.scatter(g_r, r_i, s=2)
plt.scatter(first_set_g_r, first_set_r_i, s=10, c='r')
plt.ylabel('g-r')

plt.subplot(223)
plt.hold(True)
plt.scatter(r_i, i_z, s=2)
plt.scatter(first_set_r_i, first_set_i_z, s=10, c='r')
plt.ylabel('i-z')
plt.xlabel('r-i')

plt.subplot(224)
plt.hold(True)
plt.scatter(r_i, i, s=2)
plt.scatter(first_set_r_i, first_set_i, s=10, c='r')
plt.ylabel('i')
plt.xlabel('r-i')

plt.show()
```
