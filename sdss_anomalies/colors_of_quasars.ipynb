# Colors of Quasars

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('../quasar_colors_rahul.poruri.csv')

redshift = np.zeros(len(data))
u, g, r, i, z = np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data))
u_g, g_r, r_i, i_z = np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data)), np.zeros(len(data))
```

```python
redshift = data['z']
u, g, r, i, z = data['modelMag_u'], data['modelMag_g'], data['modelMag_r'], data['modelMag_i'], data['modelMag_z']
```

```python
for item in np.where(u==-9999.0):
    u[item] = np.mean(u)
    
for item in np.where(g==-9999.0):
    g[item] = np.mean(g)

for item in np.where(r==-9999.0):
    r[item] = np.mean(r)

for item in np.where(i==-9999.0):
    i[item] = np.mean(i)
    
for item in np.where(z==-9999.0):
    z[item] = np.mean(z)
```

```python
u_g, g_r, r_i, i_z = data['modelMag_u'] - data['modelMag_g'], data['modelMag_g'] - data['modelMag_r'], data['modelMag_r'] - data['modelMag_i'], data['modelMag_i'] - data['modelMag_z']
```

```python
plt.hist(redshift, bins=50)
```

```python
plt.ylim(-4,5)
plt.scatter(redshift, u_g, s=4)
plt.ylabel('u-g')
plt.xlabel('redshift')
```

```python
plt.subplot(221)
plt.ylim(-4,5)
plt.scatter(redshift, u_g, s=4)
plt.ylabel('u-g')

plt.subplot(222)
plt.ylim(-4,5)
plt.scatter(redshift, g_r, s=4)
plt.ylabel('g-r')

plt.subplot(223)
plt.ylim(-4,5)
plt.scatter(redshift, r_i, s=4)
plt.ylabel('r-i')
plt.xlabel('redshift')

plt.subplot(224)
plt.ylim(-4,5)
plt.scatter(redshift, i_z, s=4)
plt.ylabel('i-z')
plt.xlabel('redshift')

plt.show()
```

```python
plt.subplot(221)
plt.scatter(u_g, g_r, s=2)
plt.ylabel('u-g')

plt.subplot(222)
plt.scatter(g_r, r_i, s=2)
plt.ylabel('g-r')

plt.subplot(223)
plt.scatter(r_i, i_z, s=2)
plt.ylabel('i-z')
plt.xlabel('r-i')

plt.subplot(224)
plt.scatter(r_i, i, s=2)
plt.ylabel('i')
plt.xlabel('r-i')

plt.show()
```
